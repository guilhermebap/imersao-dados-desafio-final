### Importando as bibliotecas necessárias e o dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.color_palette("rocket", as_cmap=True)
sns.set()
```

```python
url_dados = 'https://github.com/guilhermebap/imersao-dados-desafio-final/blob/main/Dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(url_dados, compression='zip')

dados.head()
```

```python
url_dados = 'https://github.com/guilhermebap/imersao-dados-desafio-final/blob/main/Dados/dados_resultados.csv?raw=true'
dados_resultados = pd.read_csv(url_dados)

dados_resultados.head()
```

```python
dados_combinados = pd.merge(dados, dados_resultados, on='id')
dados_combinados.head()
```

### Analise exploratório dos dados

```python
print('Dados Experimento: ')
print(f'O tamanho do dataset é: {dados.shape}')
print(f'A quantidade de dados nulos é: {dados.isna().sum().sum()}')
```

```python
print('Dados Resultado: ')
print(f'O tamanho do dataset é: {dados_resultados.shape}')
print(f'A quantidade de dados nulos é: {dados_resultados.isna().sum().sum()}')
```

```python
print('Dados combinados: ')
print(f'O tamanho do dataset é: {dados_combinados.shape}')
print(f'A quantidade de dados nulos é: {dados_combinados.isna().sum().sum()}')
```

```python
for x in dados_combinados.columns[1:4]:
    print(dados_combinados[x].value_counts(normalize=True))
    print()
```

```python
print(f'Número de drogas diferentes testadas: {dados_combinados["droga"].nunique()}')
```

```python
df_moa = pd.DataFrame(dados_combinados.iloc[:, 877:].sum(axis=0).sort_values(ascending=False)).reset_index()
df_moa.columns = ['moa', 'quantidade']

fig, ax = plt.subplots(figsize=(15,8))
ax.bar(df_moa.loc[:9, 'moa'], df_moa.loc[:9, 'quantidade'])
ax.set_title('Mecanismos de ação (MOA) mais frequentes', fontsize=18)
ax.set_xlabel('MOA', fontsize=14)
ax.set_ylabel('Quantidade', fontsize=14)

ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(labels=df_moa['moa'][:10].values, rotation=45, fontsize=12)

plt.show()
```

```python
print('Observações que mais tiveram MOA ativos ao mesmo tempo:')
dados_combinados.iloc[:, 877:].sum(axis=1).sort_values(ascending=False)[:20]
```

```python
principais_drogas = dados_combinados['droga'].value_counts().index[:5]

fig = plt.figure(figsize=(12,8))
ax = sns.countplot(x='droga', data=dados_combinados.query('droga in @principais_drogas'), order=principais_drogas)
ax.set_title('Gráfico de frequência das 5 drogas mais frequêntes', fontsize=18)
ax.set_xlabel('Droga', fontsize=14)
ax.set_ylabel('Quantidade', fontsize=14)
plt.show()
```

```python
ax = sns.displot(x='g-0', data=dados_combinados, aspect=1, height=5, col='dose', row='tempo', hue='tratamento')
plt.title('Gráfico de frequência do gene g-0 para tempo e dose')

plt.show()
```

```python
ax = sns.displot(x='g-1', data=dados_combinados, aspect=1, height=5, col='dose', row='tempo', hue='tratamento')
plt.title('Gráfico de frequência do gene g-1 para tempo e dose')

plt.show()
```

```python
ax = sns.displot(x='c-0', data=dados_combinados, aspect=1, height=5, col='dose', row='tempo', hue='tratamento')
plt.title('Gráfico de frequência do gene c-0 para tempo e dose')

plt.show()
```

```python
ax = sns.displot(x='g-0', data=dados_combinados[dados_combinados[df_moa['moa'][0]] == 1], aspect=1, height=5, 
                 col='dose', row='tempo')
plt.suptitle(f'Gráfico de frequência do gene g-0 nos casos onde o MOA nfkb_inhibitor está ativo para tempo e dose', y=1.01)
plt.show()
```

```python
ax = sns.displot(x='c-0', data=dados_combinados[dados_combinados[df_moa['moa'][0]] == 1], aspect=1, height=5, 
                 col='dose', row='tempo')
plt.suptitle(f'Gráfico de frequência do gene c-0 nos casos onde o MOA nfkb_inhibitor está ativo para tempo e dose', y=1.01)
plt.show()
```
